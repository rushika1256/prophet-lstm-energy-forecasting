```python
# ================================================================
# ✅ Corrected Robust Feature-Augmented Hybrid (Prophet + LSTM)
# Includes: CPI, SHAP Stability (surrogate TreeExplainer), Adversarial AUC, PRR per zone, and full plots
# ================================================================

import os
import numpy as np
import pandas as pd
import matplotlib
matplotlib.use('Agg')  # disable plot rendering (no figures shown in Colab)
import matplotlib.pyplot as plt
from prophet import Prophet
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import mean_squared_error, r2_score, roc_auc_score
from sklearn.model_selection import TimeSeriesSplit
from sklearn.ensemble import GradientBoostingClassifier, GradientBoostingRegressor
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, GaussianNoise
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.layers import BatchNormalization
import shap
import random
import warnings

warnings.filterwarnings("ignore")


# ----------------------- Config -----------------------
CSV_PATH = "Boujdour_hourly_mean.csv"   # path to your CSV
DATETIME_COL = "DateTime"
TARGET_ZONES = ["zone1", "zone2", "zone3"]   # adjust to your columns
LAGS = 24
WINDOW = 24
TEST_FRAC = 0.15
SEED = 42
os.makedirs("models", exist_ok=True)

np.random.seed(SEED)
random.seed(SEED)

# ----------------------- Helper Functions -----------------------
def fit_prophet_components(series):
    """Fit Prophet and return dataframe with trend/weekly/yearly indexed by DateTime."""
    dfp = series.reset_index().rename(columns={DATETIME_COL: "ds", series.name: "y"})
    # m = Prophet(
    #     yearly_seasonality=True,
    #     weekly_seasonality=True,
    #     seasonality_mode="multiplicative",
    #     changepoint_prior_scale=0.3
    # )
    m = Prophet(
      yearly_seasonality=True,
      weekly_seasonality=True,
      daily_seasonality=True,          # capture daily cycles
      seasonality_mode='additive',     # less multiplicative coupling
      changepoint_prior_scale=0.5      # allows smoother trend changes
    )

    m.fit(dfp)
    forecast = m.predict(dfp[["ds"]])
    comps = forecast[[c for c in ["trend", "weekly", "yearly"] if c in forecast.columns]].copy()
    comps[DATETIME_COL] = dfp["ds"].values
    comps = comps.set_index(DATETIME_COL)
    comps.index = pd.to_datetime(comps.index)
    return comps, m

def create_lag_features(df, target_col, lags):
    for lag in range(1, lags + 1):
        df[f"{target_col}_lag{lag}"] = df[target_col].shift(lag)
    return df

def create_sequences(data: np.ndarray, window: int):
    X, y = [], []
    for i in range(window, len(data)):
        X.append(data[i - window:i, :])
        y.append(data[i, 0])
    X = np.array(X)
    y = np.array(y)
    return X, y

# New: SHAP stability using surrogate tree
def compute_shap_stability_surrogate(model, X_train, WINDOW, n_features, n_splits=3,
                                     bg_samples=200, val_samples=100, random_state=SEED):
    """
    Approximate LSTM with a tree regressor on flattened inputs, then use TreeExplainer.
    Returns mean correlation between consecutive folds' mean-abs SHAP vectors.
    Fallback to perturbation if sample sizes are tiny.
    """
    tscv = TimeSeriesSplit(n_splits=n_splits)
    sims = []
    prev = None
    fold_count = 0

    for tr_idx, val_idx in tscv.split(X_train):
        fold_count += 1
        X_tr = X_train[tr_idx]
        X_val = X_train[val_idx]

        # flatten
        X_tr_flat = X_tr.reshape(X_tr.shape[0], -1)
        X_val_flat = X_val.reshape(X_val.shape[0], -1)

        # subsample background and val to manageable sizes
        if X_tr_flat.shape[0] < 10:
            # not enough samples to fit surrogate — fallback
            break

        bg_n = min(bg_samples, X_tr_flat.shape[0])
        val_n = min(val_samples, X_val_flat.shape[0])

        X_bg = X_tr_flat[:bg_n]
        X_val_sub = X_val_flat[:val_n]

        # target for surrogate: LSTM predictions on flattened bg
        try:
            y_bg = model.predict(X_bg.reshape(-1, WINDOW, n_features), verbose=0).ravel()
            # train small tree surrogate
            surrogate = GradientBoostingRegressor(n_estimators=200, max_depth=4, random_state=random_state)
            surrogate.fit(X_bg, y_bg)

            # SHAP TreeExplainer on surrogate
            explainer = shap.TreeExplainer(surrogate)
            shap_vals = explainer.shap_values(X_val_sub)  # shape (n_val, n_features_flat)
            shap_mean = np.mean(np.abs(shap_vals), axis=0)

        except Exception as e:
            # Fallback perturbation importance (cheap)
            print("Surrogate SHAP failed on fold (fallback). Reason:", str(e))
            base_preds = model.predict(X_bg[:min(20, X_bg.shape[0])].reshape(-1, WINDOW, n_features), verbose=0).ravel()
            base_mean = np.mean(base_preds)
            shap_mean = np.zeros(X_bg.shape[1])
            subset = X_bg[:min(10, X_bg.shape[0])]
            for f in range(X_bg.shape[1]):
                pert = subset.copy()
                pert[:, f] += 1e-3
                pert_preds = model.predict(pert.reshape(-1, WINDOW, n_features), verbose=0).ravel()
                shap_mean[f] = np.mean(np.abs(pert_preds - base_mean))

        # correlate with prev
        if prev is not None:
            if np.all(np.isfinite(prev)) and np.all(np.isfinite(shap_mean)):
                corr = np.corrcoef(prev, shap_mean)[0, 1]
                sims.append(corr)
        prev = shap_mean

    if len(sims) == 0:
        return np.nan
    return float(np.nanmean(sims))

# ----------------------- NEW: PRR function -----------------------
def compute_prr(y_true, y_pred):
    """
    PRR = 1 - |corr(predictions, residuals)|
    Residuals = y_true - y_pred
    If correlation is NaN, return 0.
    """
    residuals = y_true - y_pred
    try:
        corr = np.corrcoef(y_pred, residuals)[0, 1]
        prr = 1 - abs(corr)
        if np.isnan(prr):
            prr = 0.0
    except Exception:
        prr = 0.0
    return float(prr)

# ----------------------- Load CSV -----------------------
if not os.path.exists(CSV_PATH):
    raise FileNotFoundError(f"{CSV_PATH} not found. Please create it first.")

df = pd.read_csv(CSV_PATH, sep=';')
# handle case where datetime column was saved as unnamed index
if DATETIME_COL not in df.columns:
    df.rename(columns={df.columns[0]: DATETIME_COL}, inplace=True)
df[DATETIME_COL] = pd.to_datetime(df[DATETIME_COL])
df = df.set_index(DATETIME_COL).sort_index()

# Identify zones present
zones = [z for z in TARGET_ZONES if z in df.columns]
if len(zones) == 0:
    raise ValueError("No target zones found in CSV. Check TARGET_ZONES and CSV columns.")

summary_results = []

# ----------------------- Per-zone pipeline -----------------------
for zone in zones:
    print("\n" + "="*70)
    print(f"Processing zone: {zone}")
    print("="*70)

    # 1) Prepare series and Prophet components
    ts = df[[zone]].asfreq("1h").ffill().bfill()
    comps, _ = fit_prophet_components(ts[zone])
    data = ts.join(comps, how='left').fillna(0.0)

    # 2) lag features
    data = create_lag_features(data, zone, LAGS)

    # Drop rows with NaNs (from lags)
    data = data.dropna()
    if len(data) < WINDOW + 20:
        raise ValueError(f"Not enough data for zone {zone} after lagging/windowing. Need more rows.")

    # 3) Define features and target
    lag_cols = [f"{zone}_lag{l}" for l in range(1, LAGS+1)]
    prophet_cols = [c for c in data.columns if c in ["trend", "weekly", "yearly"]]
    feature_cols = prophet_cols + lag_cols
    target_col = zone

    # 4) convert to numpy and scale
    X_df = data[feature_cols].astype(float)
    y_df = data[[target_col]].astype(float)

    scaler_X = RobustScaler()
    scaler_y = RobustScaler()
    X_scaled = scaler_X.fit_transform(X_df.values)
    y_scaled = scaler_y.fit_transform(y_df.values)  # shape (n,1)

    # 5) combine target + features to create sequences (keeping target as first column)
    combined = np.hstack([y_scaled, X_scaled])  # first col is scaled target
    # remove rows with NaN/Infs just in case
    mask = np.isfinite(combined).all(axis=1)
    combined = combined[mask]
    if combined.shape[0] < WINDOW + 20:
        raise ValueError(f"After NaN cleanup, not enough rows for zone {zone}.")

    X_seq, y_seq = create_sequences(combined, WINDOW)  # X_seq shape (n_seq, WINDOW, n_features); y_seq shape (n_seq,)
    # small gaussian augmentation (optional)
    X_seq = X_seq + np.random.normal(0, 1e-4, X_seq.shape)

    # 6) Train/test split
    split_idx = int(len(X_seq) * (1 - TEST_FRAC))
    X_train, X_test = X_seq[:split_idx], X_seq[split_idx:]
    y_train, y_test = y_seq[:split_idx], y_seq[split_idx:]

    print(f"Sequences: total={len(X_seq)}, train={len(X_train)}, test={len(X_test)}")

    # # 7) Build LSTM model per zone (stable but expressive)
    # n_features = X_seq.shape[2]
    # model = Sequential([
    #     GaussianNoise(0.02, input_shape=(WINDOW, n_features)),
    #     LSTM(128, return_sequences=True, dropout=0.25, recurrent_dropout=0.15, kernel_regularizer=l2(1e-4)),
    #     LSTM(64, return_sequences=False, dropout=0.25, kernel_regularizer=l2(1e-4)),
    #     Dense(32, activation="relu"),
    #     Dropout(0.2),
    #     Dense(1)
    # ])
    # model.compile(optimizer="adam", loss="mse")
    # model = Sequential([
    #     GaussianNoise(0.03, input_shape=(WINDOW, n_features)),
    #     LSTM(128, return_sequences=True, dropout=0.3, recurrent_dropout=0.2, kernel_regularizer=l2(1e-3)),
    #     LSTM(64, return_sequences=False, dropout=0.3, kernel_regularizer=l2(1e-3)),
    #     Dense(32, activation="relu"),
    #     Dropout(0.3),
    #     Dense(1)
    # ])
    # model.compile(optimizer="adam", loss="mse")


    # # callbacks
    # model_file = os.path.join("models", f"best_model_{zone}.h5")
    # es = EarlyStopping(monitor="val_loss", patience=10, restore_best_weights=True)
    # mc = ModelCheckpoint(model_file, monitor="val_loss", save_best_only=True)
    # rl = ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=5, min_lr=1e-6)

    # # train
    # history = model.fit(
    #     X_train, y_train,
    #     validation_data=(X_test, y_test),
    #     epochs=20,
    #     batch_size=16,
    #     callbacks=[es, mc, rl],
    #     verbose=2
    # )

    # 3 ver----------------------- Build improved LSTM model per zone -----------------------
    n_features = X_seq.shape[2]

    model = Sequential([
        # Slightly higher Gaussian noise to decorrelate residuals
        GaussianNoise(0.05, input_shape=(WINDOW, n_features)),

        # First LSTM layer
        LSTM(128, return_sequences=True,
            dropout=0.35, recurrent_dropout=0.2, kernel_regularizer=l2(1e-4)),

        # Second LSTM layer
        LSTM(64, return_sequences=False,
            dropout=0.3, kernel_regularizer=l2(1e-4)),

        # Dense block with batch normalization to stabilize outputs
        Dense(32, activation="relu"),
        BatchNormalization(),
        Dropout(0.25),

        # Final output
        Dense(1)
    ])

    # Compile model
    model.compile(optimizer="adam", loss="mse")

    # Callbacks
    model_file = os.path.join("models", f"best_model_{zone}.h5")
    es = EarlyStopping(monitor="val_loss", patience=15, restore_best_weights=True)
    mc = ModelCheckpoint(model_file, monitor="val_loss", save_best_only=True)
    rl = ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=5, min_lr=1e-6)

    # Train
    history = model.fit(
        X_train, y_train,
        validation_data=(X_test, y_test),
        epochs=20,             # slightly higher epochs for stable convergence
        batch_size=16,         # smaller batch improves generalization
        callbacks=[es, mc, rl],
        verbose=2
    )

    # 8) Predictions and inverse transform
    pred_scaled = model.predict(X_test, verbose=0).reshape(-1, 1)  # shape (n_test,1)
    if not np.isfinite(pred_scaled).all():
        raise ValueError("Model produced NaN or Inf predictions. Check data and model stability.")

    pred = scaler_y.inverse_transform(pred_scaled)[:, 0]  # shape (n_test,)
    y_true = scaler_y.inverse_transform(y_test.reshape(-1, 1))[:, 0]

    # 9) metrics
    rmse = np.sqrt(mean_squared_error(y_true, pred))
    r2 = r2_score(y_true, pred)
    print(f"[{zone}] RMSE={rmse:.4f}, R2={r2:.4f}")

    # ----------------------- NEW: PRR per zone -----------------------
    prr = compute_prr(y_true, pred)
    print(f"[{zone}] PRR = {prr:.4f}")

    # 10) CPI (95%)
    residuals = np.abs(y_true - pred)
    q95 = np.quantile(residuals, 0.95)
    lower = pred - q95
    upper = pred + q95
    coverage = np.mean((y_true >= lower) & (y_true <= upper))
    width = float(np.mean(upper - lower))
    print(f"[{zone}] CPI coverage(95%) = {coverage:.3f}, avg width = {width:.4f}")

    # 11) SHAP Stability via surrogate TreeExplainer (robust)
    # shap_idx = compute_shap_stability_surrogate(model, X_train, WINDOW, n_features,
    #                                             n_splits=3, bg_samples=200, val_samples=100)
    # print(f"[{zone}] SHAP Stability Index (surrogate): {shap_idx}")

   # 11) SHAP Stability skipped to save time
    shap_idx = np.nan
    print(f"[{zone}] SHAP Stability skipped.")


    # 12) Adversarial validation (train vs test)
    def adversarial_validation(X_train, X_test):
        X_tr = X_train.reshape(X_train.shape[0], -1)
        X_te = X_test.reshape(X_test.shape[0], -1)
        X_all = np.vstack([X_tr, X_te])
        y_all = np.hstack([np.zeros(len(X_tr)), np.ones(len(X_te))])
        clf = GradientBoostingClassifier(n_estimators=150, max_depth=3, random_state=SEED)
        clf.fit(X_all, y_all)
        probs = clf.predict_proba(X_all)[:, 1]
        auc = roc_auc_score(y_all, probs)
        return auc, clf

    # adv_auc, adv_clf = adversarial_validation(X_train, X_test)
    # print(f"[{zone}] Adversarial Validation AUC = {adv_auc:.4f}")

    adv_auc = np.nan
    print(f"[{zone}] Adversarial validation skipped.")

    # ----------------------- Plotting -----------------------
    # training loss
    plt.figure(figsize=(8,4))
    plt.plot(history.history.get('loss', []), label='train_loss')
    plt.plot(history.history.get('val_loss', []), label='val_loss')
    plt.title(f"{zone} - Training Loss"); plt.legend(); plt.show()

    # Actual vs Predicted with CPI band (last N)
    plt.figure(figsize=(12,4))
    show_n = min(300, len(y_true))
    idx = np.arange(len(y_true))[-show_n:]
    plt.plot(idx, y_true[-show_n:], label='Actual', color='black')
    plt.plot(idx, pred[-show_n:], label='Pred', color='tab:blue')
    plt.fill_between(idx, lower[-show_n:], upper[-show_n:], color='gray', alpha=0.3, label='CPI 95%')
    plt.title(f"{zone} - Actual vs Pred (last {show_n})"); plt.legend(); plt.show()

    # Scatter true vs pred
    plt.figure(figsize=(6,5))
    plt.scatter(y_true, pred, alpha=0.4)
    mn, mx = np.min([y_true.min(), pred.min()]), np.max([y_true.max(), pred.max()])
    plt.plot([mn, mx], [mn, mx], color='red', linestyle='--')
    plt.xlabel("True"); plt.ylabel("Predicted")
    plt.title(f"{zone} - True vs Pred (R2={r2:.4f})"); plt.show()

    # Residual histogram
    plt.figure(figsize=(8,3))
    plt.hist(y_true - pred, bins=40, color='gray')
    plt.title(f"{zone} - Residual Distribution (mean={np.mean(y_true-pred):.4f})"); plt.show()

    # Adversarial ROC curve
    try:
        from sklearn.metrics import roc_curve, auc
        X_tr = X_train.reshape(X_train.shape[0], -1)
        X_te = X_test.reshape(X_test.shape[0], -1)
        X_all = np.vstack([X_tr, X_te])
        y_all = np.hstack([np.zeros(len(X_tr)), np.ones(len(X_te))])
        probs = adv_clf.predict_proba(X_all)[:,1]
        fpr, tpr, _ = roc_curve(y_all, probs)
        roc_auc = auc(fpr, tpr)
        plt.figure(figsize=(6,5))
        plt.plot(fpr, tpr, label=f"AUC={roc_auc:.3f}")
        plt.plot([0,1],[0,1],'--', color='gray')
        plt.title(f"{zone} - Adversarial ROC"); plt.xlabel("FPR"); plt.ylabel("TPR"); plt.legend(); plt.show()
    except Exception as e:
        print("Could not plot adversarial ROC:", e)

    # Save results summary (including PRR)
    summary_results.append({
        "Zone": zone,
        "RMSE": float(rmse),
        "R2": float(r2),
        "PRR": float(prr),
        "CPI_Coverage": float(coverage),
        "CPI_Width": float(width),
        "SHAP_Stability": float(shap_idx) if shap_idx is not None else None,
        "Adversarial_AUC": float(adv_auc)
    })

# ----------------------- Final summary table -----------------------
res_df = pd.DataFrame(summary_results)
print("\n✅ Final Results Summary:")
print(res_df.round(4))

# Combined barplots (R2, Adversarial AUC, PRR)
plt.figure(figsize=(12,4))
plt.subplot(1,3,1); plt.bar(res_df['Zone'], res_df['R2'], color='skyblue'); plt.ylim(0,1.05); plt.title("R² per Zone")
plt.subplot(1,3,2); plt.bar(res_df['Zone'], res_df['Adversarial_AUC'], color='salmon'); plt.ylim(0,1.05); plt.title("Adversarial AUC per Zone")
plt.subplot(1,3,3); plt.bar(res_df['Zone'], res_df['PRR'], color='mediumpurple'); plt.ylim(0,1.05); plt.title("PRR per Zone")
plt.tight_layout(); plt.show()
```

```python
# ================================================================
# ✅ Simplified Robust Feature-Augmented Hybrid (Prophet + LSTM)
# Keeps only final "Actual vs Predicted" (orange) plot
# ================================================================

import os
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from prophet import Prophet
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import mean_squared_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, GaussianNoise, BatchNormalization
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
import random
import warnings
%matplotlib inline

warnings.filterwarnings("ignore")

# ----------------------- Config -----------------------
CSV_PATH = "Boujdour_hourly_mean.csv"
DATETIME_COL = "DateTime"
TARGET_ZONES = ["zone1", "zone2", "zone3"]
LAGS = 24
WINDOW = 24
TEST_FRAC = 0.15
SEED = 42
os.makedirs("models", exist_ok=True)

np.random.seed(SEED)
random.seed(SEED)

# ----------------------- Helper Functions -----------------------
def fit_prophet_components(series):
    """Fit Prophet and return dataframe with trend/weekly/yearly indexed by DateTime."""
    dfp = series.reset_index().rename(columns={DATETIME_COL: "ds", series.name: "y"})
    m = Prophet(
        yearly_seasonality=True,
        weekly_seasonality=True,
        daily_seasonality=True,
        seasonality_mode='additive',
        changepoint_prior_scale=0.5
    )
    m.fit(dfp)
    forecast = m.predict(dfp[["ds"]])
    comps = forecast[[c for c in ["trend", "weekly", "yearly"] if c in forecast.columns]].copy()
    comps[DATETIME_COL] = dfp["ds"].values
    comps = comps.set_index(DATETIME_COL)
    comps.index = pd.to_datetime(comps.index)
    return comps, m

def create_lag_features(df, target_col, lags):
    for lag in range(1, lags + 1):
        df[f"{target_col}_lag{lag}"] = df[target_col].shift(lag)
    return df

def create_sequences(data: np.ndarray, window: int):
    X, y = [], []
    for i in range(window, len(data)):
        X.append(data[i - window:i, :])
        y.append(data[i, 0])
    return np.array(X), np.array(y)

def compute_prr(y_true, y_pred):
    """PRR = 1 - |corr(predictions, residuals)|"""
    residuals = y_true - y_pred
    try:
        corr = np.corrcoef(y_pred, residuals)[0, 1]
        prr = 1 - abs(corr)
        if np.isnan(prr):
            prr = 0.0
    except Exception:
        prr = 0.0
    return float(prr)

# ----------------------- Load CSV -----------------------
if not os.path.exists(CSV_PATH):
    raise FileNotFoundError(f"{CSV_PATH} not found. Please create it first.")

df = pd.read_csv(CSV_PATH, sep=';')
if DATETIME_COL not in df.columns:
    df.rename(columns={df.columns[0]: DATETIME_COL}, inplace=True)
df[DATETIME_COL] = pd.to_datetime(df[DATETIME_COL])
df = df.set_index(DATETIME_COL).sort_index()

zones = [z for z in TARGET_ZONES if z in df.columns]
if len(zones) == 0:
    raise ValueError("No target zones found in CSV. Check TARGET_ZONES and CSV columns.")

summary_results = []

# ----------------------- Per-zone pipeline -----------------------
for zone in zones:
    print("\n" + "="*70)
    print(f"Processing zone: {zone}")
    print("="*70)

    # 1) Prophet components
    ts = df[[zone]].asfreq("1h").ffill().bfill()
    comps, _ = fit_prophet_components(ts[zone])
    data = ts.join(comps, how='left').fillna(0.0)

    # 2) Lag features
    data = create_lag_features(data, zone, LAGS)
    data = data.dropna()
    if len(data) < WINDOW + 20:
        raise ValueError(f"Not enough data for zone {zone} after lagging/windowing.")

    # 3) Features + target
    lag_cols = [f"{zone}_lag{l}" for l in range(1, LAGS+1)]
    prophet_cols = [c for c in data.columns if c in ["trend", "weekly", "yearly"]]
    feature_cols = prophet_cols + lag_cols
    target_col = zone

    X_df = data[feature_cols].astype(float)
    y_df = data[[target_col]].astype(float)

    scaler_X = RobustScaler()
    scaler_y = RobustScaler()
    X_scaled = scaler_X.fit_transform(X_df.values)
    y_scaled = scaler_y.fit_transform(y_df.values)

    combined = np.hstack([y_scaled, X_scaled])
    mask = np.isfinite(combined).all(axis=1)
    combined = combined[mask]
    if combined.shape[0] < WINDOW + 20:
        raise ValueError(f"After NaN cleanup, not enough rows for zone {zone}.")

    X_seq, y_seq = create_sequences(combined, WINDOW)
    X_seq = X_seq + np.random.normal(0, 1e-4, X_seq.shape)

    split_idx = int(len(X_seq) * (1 - TEST_FRAC))
    X_train, X_test = X_seq[:split_idx], X_seq[split_idx:]
    y_train, y_test = y_seq[:split_idx], y_seq[split_idx:]

    print(f"Sequences: total={len(X_seq)}, train={len(X_train)}, test={len(X_test)}")

    # 4) LSTM Model
    n_features = X_seq.shape[2]
    model = Sequential([
        GaussianNoise(0.05, input_shape=(WINDOW, n_features)),
        LSTM(128, return_sequences=True, dropout=0.35, recurrent_dropout=0.2, kernel_regularizer=l2(1e-4)),
        LSTM(64, return_sequences=False, dropout=0.3, kernel_regularizer=l2(1e-4)),
        Dense(32, activation="relu"),
        BatchNormalization(),
        Dropout(0.25),
        Dense(1)
    ])
    model.compile(optimizer="adam", loss="mse")

    model_file = os.path.join("models", f"best_model_{zone}.h5")
    es = EarlyStopping(monitor="val_loss", patience=15, restore_best_weights=True)
    mc = ModelCheckpoint(model_file, monitor="val_loss", save_best_only=True)
    rl = ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=5, min_lr=1e-6)

    history = model.fit(
        X_train, y_train,
        validation_data=(X_test, y_test),
        epochs=20,
        batch_size=16,
        callbacks=[es, mc, rl],
        verbose=2
    )

    # 5) Predictions
    pred_scaled = model.predict(X_test, verbose=0).reshape(-1, 1)
    if not np.isfinite(pred_scaled).all():
        raise ValueError("Model produced NaN/Inf predictions.")
    pred = scaler_y.inverse_transform(pred_scaled)[:, 0]
    y_true = scaler_y.inverse_transform(y_test.reshape(-1, 1))[:, 0]

    # 6) Metrics
    rmse = np.sqrt(mean_squared_error(y_true, pred))
    r2 = r2_score(y_true, pred)
    prr = compute_prr(y_true, pred)

    residuals = np.abs(y_true - pred)
    q95 = np.quantile(residuals, 0.95)
    lower, upper = pred - q95, pred + q95
    coverage = np.mean((y_true >= lower) & (y_true <= upper))
    width = float(np.mean(upper - lower))

    print(f"[{zone}] RMSE={rmse:.4f}, R2={r2:.4f}, PRR={prr:.4f}")
    print(f"[{zone}] CPI coverage(95%)={coverage:.3f}, avg width={width:.4f}")

    # ----------------------- Only Keep Final Plot -----------------------
    plt.figure(figsize=(12, 5))
    plt.plot(data.index[-len(y_true):], y_true, label='Actual', color='black')
    plt.plot(data.index[-len(y_true):], pred, label='LSTM + Prophet', color='orange', linewidth=2)
    plt.title(f'{zone} | RMSE: {rmse:.2f}, R²: {r2:.2f}')
    plt.xlabel('Time')
    plt.ylabel('Current Consumption')
    plt.legend()
    plt.tight_layout()
    plt.show()


    summary_results.append({
        "Zone": zone,
        "RMSE": float(rmse),
        "R2": float(r2),
        "PRR": float(prr),
        "CPI_Coverage": float(coverage),
        "CPI_Width": float(width)
    })

# ----------------------- Final Summary -----------------------
res_df = pd.DataFrame(summary_results)
print("\n✅ Final Results Summary:")
print(res_df.round(4))
```

```python
# ================================================================
# Includes: CPI, SHAP Stability (surrogate TreeExplainer), Adversarial AUC, PRR per zone, and full plots
# ================================================================

import os
import numpy as np
import pandas as pd
import matplotlib
matplotlib.use('TkAgg')
import matplotlib.pyplot as plt
from prophet import Prophet
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import mean_squared_error, r2_score, roc_auc_score
from sklearn.model_selection import TimeSeriesSplit
from sklearn.ensemble import GradientBoostingClassifier, GradientBoostingRegressor
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, GaussianNoise
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.layers import BatchNormalization
import shap
import random
import warnings

warnings.filterwarnings("ignore")


# ----------------------- Config -----------------------
CSV_PATH = "Boujdour_hourly_mean.csv"   # path to your CSV
DATETIME_COL = "DateTime"
TARGET_ZONES = ["zone1", "zone2", "zone3"]   # adjust to your columns
LAGS = 24
WINDOW = 24
TEST_FRAC = 0.15
SEED = 42
os.makedirs("models", exist_ok=True)

np.random.seed(SEED)
random.seed(SEED)

# ----------------------- Helper Functions -----------------------
def fit_prophet_components(series):
    """Fit Prophet and return dataframe with trend/weekly/yearly indexed by DateTime."""
    dfp = series.reset_index().rename(columns={DATETIME_COL: "ds", series.name: "y"})
    # m = Prophet(
    #     yearly_seasonality=True,
    #     weekly_seasonality=True,
    #     seasonality_mode="multiplicative",
    #     changepoint_prior_scale=0.3
    # )
    m = Prophet(
      yearly_seasonality=True,
      weekly_seasonality=True,
      daily_seasonality=True,          # capture daily cycles
      seasonality_mode='additive',     # less multiplicative coupling
      changepoint_prior_scale=0.5      # allows smoother trend changes
    )

    m.fit(dfp)
    forecast = m.predict(dfp[["ds"]])
    comps = forecast[[c for c in ["trend", "weekly", "yearly"] if c in forecast.columns]].copy()
    comps[DATETIME_COL] = dfp["ds"].values
    comps = comps.set_index(DATETIME_COL)
    comps.index = pd.to_datetime(comps.index)
    return comps, m

def create_lag_features(df, target_col, lags):
    for lag in range(1, lags + 1):
        df[f"{target_col}_lag{lag}"] = df[target_col].shift(lag)
    return df

def create_sequences(data: np.ndarray, window: int):
    X, y = [], []
    for i in range(window, len(data)):
        X.append(data[i - window:i, :])
        y.append(data[i, 0])
    X = np.array(X)
    y = np.array(y)
    return X, y

# New: SHAP stability using surrogate tree
def compute_shap_stability_surrogate(model, X_train, WINDOW, n_features, n_splits=3,
                                     bg_samples=200, val_samples=100, random_state=SEED):
    """
    Approximate LSTM with a tree regressor on flattened inputs, then use TreeExplainer.
    Returns mean correlation between consecutive folds' mean-abs SHAP vectors.
    Fallback to perturbation if sample sizes are tiny.
    """
    tscv = TimeSeriesSplit(n_splits=n_splits)
    sims = []
    prev = None
    fold_count = 0

    for tr_idx, val_idx in tscv.split(X_train):
        fold_count += 1
        X_tr = X_train[tr_idx]
        X_val = X_train[val_idx]

        # flatten
        X_tr_flat = X_tr.reshape(X_tr.shape[0], -1)
        X_val_flat = X_val.reshape(X_val.shape[0], -1)

        # subsample background and val to manageable sizes
        if X_tr_flat.shape[0] < 10:
            # not enough samples to fit surrogate — fallback
            break

        bg_n = min(bg_samples, X_tr_flat.shape[0])
        val_n = min(val_samples, X_val_flat.shape[0])

        X_bg = X_tr_flat[:bg_n]
        X_val_sub = X_val_flat[:val_n]

        # target for surrogate: LSTM predictions on flattened bg
        try:
            y_bg = model.predict(X_bg.reshape(-1, WINDOW, n_features), verbose=0).ravel()
            # train small tree surrogate
            surrogate = GradientBoostingRegressor(n_estimators=200, max_depth=4, random_state=random_state)
            surrogate.fit(X_bg, y_bg)

            # SHAP TreeExplainer on surrogate
            explainer = shap.TreeExplainer(surrogate)
            shap_vals = explainer.shap_values(X_val_sub)  # shape (n_val, n_features_flat)
            shap_mean = np.mean(np.abs(shap_vals), axis=0)

        except Exception as e:
            # Fallback perturbation importance (cheap)
            print("Surrogate SHAP failed on fold (fallback). Reason:", str(e))
            base_preds = model.predict(X_bg[:min(20, X_bg.shape[0])].reshape(-1, WINDOW, n_features), verbose=0).ravel()
            base_mean = np.mean(base_preds)
            shap_mean = np.zeros(X_bg.shape[1])
            subset = X_bg[:min(10, X_bg.shape[0])]
            for f in range(X_bg.shape[1]):
                pert = subset.copy()
                pert[:, f] += 1e-3
                pert_preds = model.predict(pert.reshape(-1, WINDOW, n_features), verbose=0).ravel()
                shap_mean[f] = np.mean(np.abs(pert_preds - base_mean))

        # correlate with prev
        if prev is not None:
            if np.all(np.isfinite(prev)) and np.all(np.isfinite(shap_mean)):
                corr = np.corrcoef(prev, shap_mean)[0, 1]
                sims.append(corr)
        prev = shap_mean

    if len(sims) == 0:
        return np.nan
    return float(np.nanmean(sims))

# ----------------------- NEW: PRR function -----------------------
def compute_prr(y_true, y_pred):
    """
    PRR = 1 - |corr(predictions, residuals)|
    Residuals = y_true - y_pred
    If correlation is NaN, return 0.
    """
    residuals = y_true - y_pred
    try:
        corr = np.corrcoef(y_pred, residuals)[0, 1]
        prr = 1 - abs(corr)
        if np.isnan(prr):
            prr = 0.0
    except Exception:
        prr = 0.0
    return float(prr)

# ----------------------- Load CSV -----------------------
if not os.path.exists(CSV_PATH):
    raise FileNotFoundError(f"{CSV_PATH} not found. Please create it first.")

df = pd.read_csv(CSV_PATH, sep=';')
# handle case where datetime column was saved as unnamed index
if DATETIME_COL not in df.columns:
    df.rename(columns={df.columns[0]: DATETIME_COL}, inplace=True)
df[DATETIME_COL] = pd.to_datetime(df[DATETIME_COL])
df = df.set_index(DATETIME_COL).sort_index()

# Identify zones present
zones = [z for z in TARGET_ZONES if z in df.columns]
if len(zones) == 0:
    raise ValueError("No target zones found in CSV. Check TARGET_ZONES and CSV columns.")

summary_results = []

# ----------------------- Per-zone pipeline -----------------------
for zone in zones:
    print("\n" + "="*70)
    print(f"Processing zone: {zone}")
    print("="*70)

    # 1) Prepare series and Prophet components
    ts = df[[zone]].asfreq("1h").ffill().bfill()
    comps, _ = fit_prophet_components(ts[zone])
    data = ts.join(comps, how='left').fillna(0.0)

    # 2) lag features
    data = create_lag_features(data, zone, LAGS)

    # Drop rows with NaNs (from lags)
    data = data.dropna()
    if len(data) < WINDOW + 20:
        raise ValueError(f"Not enough data for zone {zone} after lagging/windowing. Need more rows.")

    # 3) Define features and target
    lag_cols = [f"{zone}_lag{l}" for l in range(1, LAGS+1)]
    prophet_cols = [c for c in data.columns if c in ["trend", "weekly", "yearly"]]
    feature_cols = prophet_cols + lag_cols
    target_col = zone

    # 4) convert to numpy and scale
    X_df = data[feature_cols].astype(float)
    y_df = data[[target_col]].astype(float)

    scaler_X = RobustScaler()
    scaler_y = RobustScaler()
    X_scaled = scaler_X.fit_transform(X_df.values)
    y_scaled = scaler_y.fit_transform(y_df.values)  # shape (n,1)

    # 5) combine target + features to create sequences (keeping target as first column)
    combined = np.hstack([y_scaled, X_scaled])  # first col is scaled target
    # remove rows with NaN/Infs just in case
    mask = np.isfinite(combined).all(axis=1)
    combined = combined[mask]
    if combined.shape[0] < WINDOW + 20:
        raise ValueError(f"After NaN cleanup, not enough rows for zone {zone}.")

    X_seq, y_seq = create_sequences(combined, WINDOW)  # X_seq shape (n_seq, WINDOW, n_features); y_seq shape (n_seq,)
    # small gaussian augmentation (optional)
    X_seq = X_seq + np.random.normal(0, 1e-4, X_seq.shape)

    # 6) Train/test split
    split_idx = int(len(X_seq) * (1 - TEST_FRAC))
    X_train, X_test = X_seq[:split_idx], X_seq[split_idx:]
    y_train, y_test = y_seq[:split_idx], y_seq[split_idx:]

    print(f"Sequences: total={len(X_seq)}, train={len(X_train)}, test={len(X_test)}")

    # # 7) Build LSTM model per zone (stable but expressive)
    # n_features = X_seq.shape[2]
    # model = Sequential([
    #     GaussianNoise(0.02, input_shape=(WINDOW, n_features)),
    #     LSTM(128, return_sequences=True, dropout=0.25, recurrent_dropout=0.15, kernel_regularizer=l2(1e-4)),
    #     LSTM(64, return_sequences=False, dropout=0.25, kernel_regularizer=l2(1e-4)),
    #     Dense(32, activation="relu"),
    #     Dropout(0.2),
    #     Dense(1)
    # ])
    # model.compile(optimizer="adam", loss="mse")
    # model = Sequential([
    #     GaussianNoise(0.03, input_shape=(WINDOW, n_features)),
    #     LSTM(128, return_sequences=True, dropout=0.3, recurrent_dropout=0.2, kernel_regularizer=l2(1e-3)),
    #     LSTM(64, return_sequences=False, dropout=0.3, kernel_regularizer=l2(1e-3)),
    #     Dense(32, activation="relu"),
    #     Dropout(0.3),
    #     Dense(1)
    # ])
    # model.compile(optimizer="adam", loss="mse")


    # # callbacks
    # model_file = os.path.join("models", f"best_model_{zone}.h5")
    # es = EarlyStopping(monitor="val_loss", patience=10, restore_best_weights=True)
    # mc = ModelCheckpoint(model_file, monitor="val_loss", save_best_only=True)
    # rl = ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=5, min_lr=1e-6)

    # # train
    # history = model.fit(
    #     X_train, y_train,
    #     validation_data=(X_test, y_test),
    #     epochs=20,
    #     batch_size=16,
    #     callbacks=[es, mc, rl],
    #     verbose=2
    # )

    # 3 ver----------------------- Build improved LSTM model per zone -----------------------
    n_features = X_seq.shape[2]

    model = Sequential([
        # Slightly higher Gaussian noise to decorrelate residuals
        GaussianNoise(0.05, input_shape=(WINDOW, n_features)),

        # First LSTM layer
        LSTM(128, return_sequences=True,
            dropout=0.35, recurrent_dropout=0.2, kernel_regularizer=l2(1e-4)),

        # Second LSTM layer
        LSTM(64, return_sequences=False,
            dropout=0.3, kernel_regularizer=l2(1e-4)),

        # Dense block with batch normalization to stabilize outputs
        Dense(32, activation="relu"),
        BatchNormalization(),
        Dropout(0.25),

        # Final output
        Dense(1)
    ])

    # Compile model
    model.compile(optimizer="adam", loss="mse")

    # Callbacks
    model_file = os.path.join("models", f"best_model_{zone}.h5")
    es = EarlyStopping(monitor="val_loss", patience=15, restore_best_weights=True)
    mc = ModelCheckpoint(model_file, monitor="val_loss", save_best_only=True)
    rl = ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=5, min_lr=1e-6)

    # Train
    history = model.fit(
        X_train, y_train,
        validation_data=(X_test, y_test),
        epochs=20,             # slightly higher epochs for stable convergence
        batch_size=16,         # smaller batch improves generalization
        callbacks=[es, mc, rl],
        verbose=2
    )

    # 8) Predictions and inverse transform
    pred_scaled = model.predict(X_test, verbose=0).reshape(-1, 1)  # shape (n_test,1)
    if not np.isfinite(pred_scaled).all():
        raise ValueError("Model produced NaN or Inf predictions. Check data and model stability.")

    pred = scaler_y.inverse_transform(pred_scaled)[:, 0]  # shape (n_test,)
    y_true = scaler_y.inverse_transform(y_test.reshape(-1, 1))[:, 0]

    # 9) metrics
    rmse = np.sqrt(mean_squared_error(y_true, pred))
    r2 = r2_score(y_true, pred)
    print(f"[{zone}] RMSE={rmse:.4f}, R2={r2:.4f}")

    # ----------------------- NEW: PRR per zone -----------------------
    prr = compute_prr(y_true, pred)
    print(f"[{zone}] PRR = {prr:.4f}")

    # 10) CPI (95%)
    residuals = np.abs(y_true - pred)
    q95 = np.quantile(residuals, 0.95)
    lower = pred - q95
    upper = pred + q95
    coverage = np.mean((y_true >= lower) & (y_true <= upper))
    width = float(np.mean(upper - lower))
    print(f"[{zone}] CPI coverage(95%) = {coverage:.3f}, avg width = {width:.4f}")

    # 11) SHAP Stability via surrogate TreeExplainer (robust)
    # shap_idx = compute_shap_stability_surrogate(model, X_train, WINDOW, n_features,
    #                                             n_splits=3, bg_samples=200, val_samples=100)
    # print(f"[{zone}] SHAP Stability Index (surrogate): {shap_idx}")

   # 11) SHAP Stability skipped to save time
    shap_idx = np.nan
    print(f"[{zone}] SHAP Stability skipped.")


    # 12) Adversarial validation (train vs test)
    def adversarial_validation(X_train, X_test):
        X_tr = X_train.reshape(X_train.shape[0], -1)
        X_te = X_test.reshape(X_test.shape[0], -1)
        X_all = np.vstack([X_tr, X_te])
        y_all = np.hstack([np.zeros(len(X_tr)), np.ones(len(X_te))])
        clf = GradientBoostingClassifier(n_estimators=150, max_depth=3, random_state=SEED)
        clf.fit(X_all, y_all)
        probs = clf.predict_proba(X_all)[:, 1]
        auc = roc_auc_score(y_all, probs)
        return auc, clf

    # adv_auc, adv_clf = adversarial_validation(X_train, X_test)
    # print(f"[{zone}] Adversarial Validation AUC = {adv_auc:.4f}")

    adv_auc = np.nan
    print(f"[{zone}] Adversarial validation skipped.")

    # ----------------------- Plotting -----------------------
    # training loss
    plt.figure(figsize=(8,4))
    plt.plot(history.history.get('loss', []), label='train_loss')
    plt.plot(history.history.get('val_loss', []), label='val_loss')
    plt.title(f"{zone} - Training Loss"); plt.legend(); plt.show()

    # Actual vs Predicted with CPI band (last N)
    plt.figure(figsize=(12,4))
    show_n = min(300, len(y_true))
    idx = np.arange(len(y_true))[-show_n:]
    plt.plot(idx, y_true[-show_n:], label='Actual', color='black')
    plt.plot(idx, pred[-show_n:], label='Pred', color='tab:blue')
    plt.fill_between(idx, lower[-show_n:], upper[-show_n:], color='gray', alpha=0.3, label='CPI 95%')
    plt.title(f"{zone} - Actual vs Pred (last {show_n})"); plt.legend(); plt.show()

    # Scatter true vs pred
    plt.figure(figsize=(6,5))
    plt.scatter(y_true, pred, alpha=0.4)
    mn, mx = np.min([y_true.min(), pred.min()]), np.max([y_true.max(), pred.max()])
    plt.plot([mn, mx], [mn, mx], color='red', linestyle='--')
    plt.xlabel("True"); plt.ylabel("Predicted")
    plt.title(f"{zone} - True vs Pred (R2={r2:.4f})"); plt.show()

    # Residual histogram
    plt.figure(figsize=(8,3))
    plt.hist(y_true - pred, bins=40, color='gray')
    plt.title(f"{zone} - Residual Distribution (mean={np.mean(y_true-pred):.4f})"); plt.show()

    # Adversarial ROC curve
    try:
        from sklearn.metrics import roc_curve, auc
        X_tr = X_train.reshape(X_train.shape[0], -1)
        X_te = X_test.reshape(X_test.shape[0], -1)
        X_all = np.vstack([X_tr, X_te])
        y_all = np.hstack([np.zeros(len(X_tr)), np.ones(len(X_te))])
        probs = adv_clf.predict_proba(X_all)[:,1]
        fpr, tpr, _ = roc_curve(y_all, probs)
        roc_auc = auc(fpr, tpr)
        plt.figure(figsize=(6,5))
        plt.plot(fpr, tpr, label=f"AUC={roc_auc:.3f}")
        plt.plot([0,1],[0,1],'--', color='gray')
        plt.title(f"{zone} - Adversarial ROC"); plt.xlabel("FPR"); plt.ylabel("TPR"); plt.legend(); plt.show()
    except Exception as e:
        print("Could not plot adversarial ROC:", e)

    # Save results summary (including PRR)
    summary_results.append({
        "Zone": zone,
        "RMSE": float(rmse),
        "R2": float(r2),
        "PRR": float(prr),
        "CPI_Coverage": float(coverage),
        "CPI_Width": float(width),
        "SHAP_Stability": float(shap_idx) if shap_idx is not None else None,
        "Adversarial_AUC": float(adv_auc)
    })

# ----------------------- Final summary table -----------------------
res_df = pd.DataFrame(summary_results)
print("\n✅ Final Results Summary:")
print(res_df.round(4))

# Combined barplots (R2, Adversarial AUC, PRR)
plt.figure(figsize=(12,4))
plt.subplot(1,3,1); plt.bar(res_df['Zone'], res_df['R2'], color='skyblue'); plt.ylim(0,1.05); plt.title("R² per Zone")
plt.subplot(1,3,2); plt.bar(res_df['Zone'], res_df['Adversarial_AUC'], color='salmon'); plt.ylim(0,1.05); plt.title("Adversarial AUC per Zone")
plt.subplot(1,3,3); plt.bar(res_df['Zone'], res_df['PRR'], color='mediumpurple'); plt.ylim(0,1.05); plt.title("PRR per Zone")
plt.tight_layout(); plt.show()

# ----------------------- Actual vs Predicted per Zone -----------------------
for zone in zones:
    print(f"\nPlotting Actual vs Predicted for {zone}...")

    # reload model and test sequences
    ts = df[[zone]].asfreq("1h").ffill().bfill()
    comps, _ = fit_prophet_components(ts[zone])
    data = ts.join(comps, how='left').fillna(0.0)
    data = create_lag_features(data, zone, LAGS).dropna()

    # Features
    lag_cols = [f"{zone}_lag{l}" for l in range(1, LAGS+1)]
    prophet_cols = [c for c in data.columns if c in ["trend", "weekly", "yearly"]]
    feature_cols = prophet_cols + lag_cols
    X_df = data[feature_cols].astype(float)
    y_df = data[[zone]].astype(float)

    # Scale
    scaler_X = RobustScaler()
    scaler_y = RobustScaler()
    X_scaled = scaler_X.fit_transform(X_df.values)
    y_scaled = scaler_y.fit_transform(y_df.values)
    combined = np.hstack([y_scaled, X_scaled])
    mask = np.isfinite(combined).all(axis=1)
    combined = combined[mask]
    X_seq, y_seq = create_sequences(combined, WINDOW)

    # Train/test split
    split_idx = int(len(X_seq) * (1 - TEST_FRAC))
    X_test_seq = X_seq[split_idx:]
    y_test_seq = y_seq[split_idx:]

    # Load trained model
    from tensorflow.keras.models import load_model
    model_file = os.path.join("models", f"best_model_{zone}.h5")
    model = load_model(model_file, compile=False)

    pred_scaled = model.predict(X_test_seq, verbose=0).reshape(-1,1)
    pred = scaler_y.inverse_transform(pred_scaled)[:,0]
    y_true = scaler_y.inverse_transform(y_test_seq.reshape(-1,1))[:,0]

    # Plot
    plt.figure(figsize=(12,4))
    show_n = min(300, len(y_true))
    idx = np.arange(len(y_true))[-show_n:]
    plt.plot(idx, y_true[-show_n:], label='Actual', color='black')
    plt.plot(idx, pred[-show_n:], label='Predicted', color='tab:blue')
    plt.title(f"{zone} - Actual vs Predicted (last {show_n})")
    plt.xlabel("Time Index")
    plt.ylabel("Power")
    plt.legend()
    plt.tight_layout()
    plt.show()

```

```python
# ================================================================
# ✅ Corrected Robust Feature-Augmented Hybrid (Prophet + LSTM) with RFE
# Includes: CPI, SHAP Stability (surrogate TreeExplainer), Adversarial AUC, PRR per zone, and full plots
# ================================================================

import os
import numpy as np
import pandas as pd
import matplotlib
matplotlib.use('Agg')  # disable plot rendering (no figures shown in Colab)
import matplotlib.pyplot as plt
from prophet import Prophet
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import mean_squared_error, r2_score, roc_auc_score
from sklearn.model_selection import TimeSeriesSplit
from sklearn.ensemble import GradientBoostingClassifier, GradientBoostingRegressor
from sklearn.feature_selection import RFE
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, GaussianNoise, BatchNormalization
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
import shap
import random
import warnings

warnings.filterwarnings("ignore")

# ----------------------- Config -----------------------
CSV_PATH = "Boujdour_hourly_mean.csv"   # path to your CSV
DATETIME_COL = "DateTime"
TARGET_ZONES = ["zone1", "zone2", "zone3"]   # adjust to your columns
LAGS = 24
WINDOW = 24
TEST_FRAC = 0.15
SEED = 42
os.makedirs("models", exist_ok=True)

np.random.seed(SEED)
random.seed(SEED)

# ----------------------- Helper Functions -----------------------
def fit_prophet_components(series):
    dfp = series.reset_index().rename(columns={DATETIME_COL: "ds", series.name: "y"})
    m = Prophet(
      yearly_seasonality=True,
      weekly_seasonality=True,
      daily_seasonality=True,
      seasonality_mode='additive',
      changepoint_prior_scale=0.5
    )
    m.fit(dfp)
    forecast = m.predict(dfp[["ds"]])
    comps = forecast[[c for c in ["trend", "weekly", "yearly"] if c in forecast.columns]].copy()
    comps[DATETIME_COL] = dfp["ds"].values
    comps = comps.set_index(DATETIME_COL)
    comps.index = pd.to_datetime(comps.index)
    return comps, m

def create_lag_features(df, target_col, lags):
    for lag in range(1, lags + 1):
        df[f"{target_col}_lag{lag}"] = df[target_col].shift(lag)
    return df

def create_sequences(data: np.ndarray, window: int):
    X, y = [], []
    for i in range(window, len(data)):
        X.append(data[i - window:i, :])
        y.append(data[i, 0])
    X = np.array(X)
    y = np.array(y)
    return X, y

def compute_prr(y_true, y_pred):
    residuals = y_true - y_pred
    try:
        corr = np.corrcoef(y_pred, residuals)[0, 1]
        prr = 1 - abs(corr)
        if np.isnan(prr):
            prr = 0.0
    except Exception:
        prr = 0.0
    return float(prr)

# ----------------------- Load CSV -----------------------
if not os.path.exists(CSV_PATH):
    raise FileNotFoundError(f"{CSV_PATH} not found. Please create it first.")

df = pd.read_csv(CSV_PATH, sep=';')
if DATETIME_COL not in df.columns:
    df.rename(columns={df.columns[0]: DATETIME_COL}, inplace=True)
df[DATETIME_COL] = pd.to_datetime(df[DATETIME_COL])
df = df.set_index(DATETIME_COL).sort_index()

# Identify zones present
zones = [z for z in TARGET_ZONES if z in df.columns]
if len(zones) == 0:
    raise ValueError("No target zones found in CSV. Check TARGET_ZONES and CSV columns.")

summary_results = []

# ----------------------- Per-zone pipeline -----------------------
for zone in zones:
    print("\n" + "="*70)
    print(f"Processing zone: {zone}")
    print("="*70)

    ts = df[[zone]].asfreq("1h").ffill().bfill()
    comps, _ = fit_prophet_components(ts[zone])
    data = ts.join(comps, how='left').fillna(0.0)
    data = create_lag_features(data, zone, LAGS)
    data = data.dropna()
    if len(data) < WINDOW + 20:
        raise ValueError(f"Not enough data for zone {zone} after lagging/windowing. Need more rows.")

    lag_cols = [f"{zone}_lag{l}" for l in range(1, LAGS+1)]
    prophet_cols = [c for c in data.columns if c in ["trend", "weekly", "yearly"]]
    feature_cols = prophet_cols + lag_cols
    target_col = zone

    X_df = data[feature_cols].astype(float)
    y_df = data[[target_col]].astype(float)

    scaler_X = RobustScaler()
    scaler_y = RobustScaler()
    X_scaled = scaler_X.fit_transform(X_df.values)
    y_scaled = scaler_y.fit_transform(y_df.values)

    # ----------------------- NEW: Recursive Feature Selection -----------------------
    X_flat = X_scaled.copy()
    y_flat = y_scaled.ravel()
    estimator = GradientBoostingRegressor(n_estimators=100, max_depth=3, random_state=SEED)
    #k_features = max(5, int(X_flat.shape[1] * 0.5))
    k_features = 20
    selector = RFE(estimator, n_features_to_select=k_features, step=1)
    selector = selector.fit(X_flat, y_flat)
    selected_cols_mask = selector.support_
    selected_feature_cols = [col for col, keep in zip(feature_cols, selected_cols_mask) if keep]
    print(f"[{zone}] RFE selected features ({len(selected_feature_cols)}): {selected_feature_cols}")
    X_scaled = X_scaled[:, selected_cols_mask]

    combined = np.hstack([y_scaled, X_scaled])
    mask = np.isfinite(combined).all(axis=1)
    combined = combined[mask]
    if combined.shape[0] < WINDOW + 20:
        raise ValueError(f"After NaN cleanup, not enough rows for zone {zone}.")

    X_seq, y_seq = create_sequences(combined, WINDOW)
    X_seq = X_seq + np.random.normal(0, 1e-4, X_seq.shape)

    split_idx = int(len(X_seq) * (1 - TEST_FRAC))
    X_train, X_test = X_seq[:split_idx], X_seq[split_idx:]
    y_train, y_test = y_seq[:split_idx], y_seq[split_idx:]

    print(f"Sequences: total={len(X_seq)}, train={len(X_train)}, test={len(X_test)}")

    n_features = X_seq.shape[2]

    model = Sequential([
        GaussianNoise(0.05, input_shape=(WINDOW, n_features)),
        LSTM(128, return_sequences=True,
             dropout=0.35, recurrent_dropout=0.2, kernel_regularizer=l2(1e-4)),
        LSTM(64, return_sequences=False,
             dropout=0.3, kernel_regularizer=l2(1e-4)),
        Dense(32, activation="relu"),
        BatchNormalization(),
        Dropout(0.25),
        Dense(1)
    ])
    model.compile(optimizer="adam", loss="mse")

    model_file = os.path.join("models", f"best_model_{zone}.h5")
    es = EarlyStopping(monitor="val_loss", patience=15, restore_best_weights=True)
    mc = ModelCheckpoint(model_file, monitor="val_loss", save_best_only=True)
    rl = ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=5, min_lr=1e-6)

    history = model.fit(
        X_train, y_train,
        validation_data=(X_test, y_test),
        epochs=20,
        batch_size=16,
        callbacks=[es, mc, rl],
        verbose=2
    )

    pred_scaled = model.predict(X_test, verbose=0).reshape(-1, 1)
    pred = scaler_y.inverse_transform(pred_scaled)[:, 0]
    y_true = scaler_y.inverse_transform(y_test.reshape(-1, 1))[:, 0]

    rmse = np.sqrt(mean_squared_error(y_true, pred))
    r2 = r2_score(y_true, pred)
    print(f"[{zone}] RMSE={rmse:.4f}, R2={r2:.4f}")

    prr = compute_prr(y_true, pred)
    print(f"[{zone}] PRR = {prr:.4f}")

    residuals = np.abs(y_true - pred)
    q95 = np.quantile(residuals, 0.95)
    lower = pred - q95
    upper = pred + q95
    coverage = np.mean((y_true >= lower) & (y_true <= upper))
    width = float(np.mean(upper - lower))
    print(f"[{zone}] CPI coverage(95%) = {coverage:.3f}, avg width = {width:.4f}")

    shap_idx = np.nan
    print(f"[{zone}] SHAP Stability skipped.")

    adv_auc = np.nan
    print(f"[{zone}] Adversarial validation skipped.")

    # ----------------------- Plotting -----------------------
    plt.figure(figsize=(8,4))
    plt.plot(history.history.get('loss', []), label='train_loss')
    plt.plot(history.history.get('val_loss', []), label='val_loss')
    plt.title(f"{zone} - Training Loss"); plt.legend(); plt.show()

    plt.figure(figsize=(12,4))
    show_n = min(300, len(y_true))
    idx = np.arange(len(y_true))[-show_n:]
    plt.plot(idx, y_true[-show_n:], label='Actual', color='black')
    plt.plot(idx, pred[-show_n:], label='Pred', color='tab:blue')
    plt.fill_between(idx, lower[-show_n:], upper[-show_n:], color='gray', alpha=0.3, label='CPI 95%')
    plt.title(f"{zone} - Actual vs Pred (last {show_n})"); plt.legend(); plt.show()

    plt.figure(figsize=(6,5))
    plt.scatter(y_true, pred, alpha=0.4)
    mn, mx = np.min([y_true.min(), pred.min()]), np.max([y_true.max(), pred.max()])
    plt.plot([mn, mx], [mn, mx], color='red', linestyle='--')
    plt.xlabel("True"); plt.ylabel("Predicted")
    plt.title(f"{zone} - True vs Pred (R2={r2:.4f})"); plt.show()

    plt.figure(figsize=(8,3))
    plt.hist(y_true - pred, bins=40, color='gray')
    plt.title(f"{zone} - Residual Distribution (mean={np.mean(y_true-pred):.4f})"); plt.show()

    summary_results.append({
        "Zone": zone,
        "RMSE": float(rmse),
        "R2": float(r2),
        "PRR": float(prr),
        "CPI_Coverage": float(coverage),
        "CPI_Width": float(width),
        "SHAP_Stability": float(shap_idx) if shap_idx is not None else None,
        "Adversarial_AUC": float(adv_auc)
    })

# ----------------------- Final summary table -----------------------
res_df = pd.DataFrame(summary_results)
print("\n✅ Final Results Summary:")
print(res_df.round(4))

plt.figure(figsize=(12,4))
plt.subplot(1,3,1); plt.bar(res_df['Zone'], res_df['R2'], color='skyblue'); plt.ylim(0,1.05); plt.title("R² per Zone")
plt.subplot(1,3,2); plt.bar(res_df['Zone'], res_df['Adversarial_AUC'], color='salmon'); plt.ylim(0,1.05); plt.title("Adversarial AUC per Zone")
plt.subplot(1,3,3); plt.bar(res_df['Zone'], res_df['PRR'], color='mediumpurple'); plt.ylim(0,1.05); plt.title("PRR per Zone")
plt.tight_layout(); plt.show()
```

```python
# ================================================================
# ✅ Robust Feature-Augmented Hybrid (Prophet + LSTM) with RFE
# SHAP removed; Actual vs Predicted uses full test set (all samples)
# ================================================================

import os
import numpy as np
import pandas as pd
import matplotlib
#matplotlib.use('Agg')  # disable plot rendering (no figures shown in Colab)
import matplotlib.pyplot as plt
from prophet import Prophet
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.feature_selection import RFE
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, GaussianNoise, BatchNormalization
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
import random
import warnings

warnings.filterwarnings("ignore")

# ----------------------- Config -----------------------
CSV_PATH = "Laayoune_hourly_mean.csv"   # path to your CSV
DATETIME_COL = "DateTime"
TARGET_ZONES = ["zone1", "zone2", "zone3", "zone4", "zone5"]   # adjust to your columns
LAGS = 24
WINDOW = 24
TEST_FRAC = 0.15
SEED = 42
os.makedirs("models", exist_ok=True)

np.random.seed(SEED)
random.seed(SEED)

# ----------------------- Helper Functions -----------------------
def fit_prophet_components(series):
    dfp = series.reset_index().rename(columns={DATETIME_COL: "ds", series.name: "y"})
    m = Prophet(
      yearly_seasonality=True,
      weekly_seasonality=True,
      daily_seasonality=True,
      seasonality_mode='additive',
      changepoint_prior_scale=0.5
    )
    m.fit(dfp)
    forecast = m.predict(dfp[["ds"]])
    comps = forecast[[c for c in ["trend", "weekly", "yearly"] if c in forecast.columns]].copy()
    comps[DATETIME_COL] = dfp["ds"].values
    comps = comps.set_index(DATETIME_COL)
    comps.index = pd.to_datetime(comps.index)
    return comps, m

def create_lag_features(df, target_col, lags):
    for lag in range(1, lags + 1):
        df[f"{target_col}_lag{lag}"] = df[target_col].shift(lag)
    return df

def create_sequences(data: np.ndarray, window: int):
    X, y = [], []
    for i in range(window, len(data)):
        X.append(data[i - window:i, :])
        y.append(data[i, 0])
    X = np.array(X)
    y = np.array(y)
    return X, y

def compute_prr(y_true, y_pred):
    residuals = y_true - y_pred
    try:
        corr = np.corrcoef(y_pred, residuals)[0, 1]
        prr = 1 - abs(corr)
        if np.isnan(prr):
            prr = 0.0
    except Exception:
        prr = 0.0
    return float(prr)

# ----------------------- Load CSV -----------------------
if not os.path.exists(CSV_PATH):
    raise FileNotFoundError(f"{CSV_PATH} not found. Please create it first.")

df = pd.read_csv(CSV_PATH, sep=';')
if DATETIME_COL not in df.columns:
    df.rename(columns={df.columns[0]: DATETIME_COL}, inplace=True)
df[DATETIME_COL] = pd.to_datetime(df[DATETIME_COL])
df = df.set_index(DATETIME_COL).sort_index()

# Identify zones present
zones = [z for z in TARGET_ZONES if z in df.columns]
if len(zones) == 0:
    raise ValueError("No target zones found in CSV. Check TARGET_ZONES and CSV columns.")

summary_results = []

# ----------------------- Per-zone pipeline -----------------------
for zone in zones:
    print("\n" + "="*70)
    print(f"Processing zone: {zone}")
    print("="*70)

    ts = df[[zone]].asfreq("1h").ffill().bfill()
    comps, _ = fit_prophet_components(ts[zone])
    data = ts.join(comps, how='left').fillna(0.0)
    data = create_lag_features(data, zone, LAGS)
    data = data.dropna()
    if len(data) < WINDOW + 20:
        raise ValueError(f"Not enough data for zone {zone} after lagging/windowing. Need more rows.")

    lag_cols = [f"{zone}_lag{l}" for l in range(1, LAGS+1)]
    prophet_cols = [c for c in data.columns if c in ["trend", "weekly", "yearly"]]
    feature_cols = prophet_cols + lag_cols
    target_col = zone

    X_df = data[feature_cols].astype(float)
    y_df = data[[target_col]].astype(float)

    scaler_X = RobustScaler()
    scaler_y = RobustScaler()
    X_scaled = scaler_X.fit_transform(X_df.values)
    y_scaled = scaler_y.fit_transform(y_df.values)

    # ----------------------- NEW: Recursive Feature Selection -----------------------
    X_flat = X_scaled.copy()
    y_flat = y_scaled.ravel()
    estimator = GradientBoostingRegressor(n_estimators=100, max_depth=3, random_state=SEED)
    k_features = 20
    selector = RFE(estimator, n_features_to_select=k_features, step=1)
    selector = selector.fit(X_flat, y_flat)
    selected_cols_mask = selector.support_
    selected_feature_cols = [col for col, keep in zip(feature_cols, selected_cols_mask) if keep]
    print(f"[{zone}] RFE selected features ({len(selected_feature_cols)}): {selected_feature_cols}")
    X_scaled = X_scaled[:, selected_cols_mask]

    # combine for sequence building, keep index alignment for plotting
    combined = np.hstack([y_scaled, X_scaled])
    mask = np.isfinite(combined).all(axis=1)
    combined = combined[mask]
    index_after_mask = data.index[mask]  # timestamps aligned to combined rows

    if combined.shape[0] < WINDOW + 20:
        raise ValueError(f"After NaN cleanup, not enough rows for zone {zone}.")

    # Build sequences and corresponding time index for each target (y)
    X_seq, y_seq = create_sequences(combined, WINDOW)
    # The y (target) timestamps correspond to index_after_mask[WINDOW:]
    y_index = index_after_mask[WINDOW:]
    if len(y_index) != len(y_seq):
        minlen = min(len(y_index), len(y_seq))
        y_index = y_index[-minlen:]
        X_seq = X_seq[-minlen:]
        y_seq = y_seq[-minlen:]

    # small noise for numeric stability
    X_seq = X_seq + np.random.normal(0, 1e-4, X_seq.shape)

    split_idx = int(len(X_seq) * (1 - TEST_FRAC))
    X_train, X_test = X_seq[:split_idx], X_seq[split_idx:]
    y_train, y_test = y_seq[:split_idx], y_seq[split_idx:]
    y_index_train = y_index[:split_idx]
    y_index_test = y_index[split_idx:]

    print(f"Sequences: total={len(X_seq)}, train={len(X_train)}, test={len(X_test)}")

    n_features = X_seq.shape[2]

    model = Sequential([
        GaussianNoise(0.05, input_shape=(WINDOW, n_features)),
        LSTM(128, return_sequences=True,
             dropout=0.35, recurrent_dropout=0.2, kernel_regularizer=l2(1e-4)),
        LSTM(64, return_sequences=False,
             dropout=0.3, kernel_regularizer=l2(1e-4)),
        Dense(32, activation="relu"),
        BatchNormalization(),
        Dropout(0.25),
        Dense(1)
    ])
    model.compile(optimizer="adam", loss="mse")

    model_file = os.path.join("models", f"best_model_{zone}.h5")
    es = EarlyStopping(monitor="val_loss", patience=15, restore_best_weights=True)
    mc = ModelCheckpoint(model_file, monitor="val_loss", save_best_only=True)
    rl = ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=5, min_lr=1e-6)

    history = model.fit(
        X_train, y_train,
        validation_data=(X_test, y_test),
        epochs=20,
        batch_size=16,
        callbacks=[es, mc, rl],
        verbose=2
    )

    pred_scaled = model.predict(X_test, verbose=0).reshape(-1, 1)
    pred = scaler_y.inverse_transform(pred_scaled)[:, 0]
    y_true = scaler_y.inverse_transform(y_test.reshape(-1, 1))[:, 0]

    # Build pandas Series with timestamps for plotting
    y_true_series = pd.Series(y_true, index=pd.to_datetime(y_index_test))
    pred_series = pd.Series(pred, index=pd.to_datetime(y_index_test))

    rmse = np.sqrt(mean_squared_error(y_true, pred))
    r2 = r2_score(y_true, pred)
    print(f"[{zone}] RMSE={rmse:.4f}, R2={r2:.4f}")

    prr = compute_prr(y_true, pred)
    print(f"[{zone}] PRR = {prr:.4f}")

    residuals = np.abs(y_true - pred)
    q95 = np.quantile(residuals, 0.95)
    lower = pred - q95
    upper = pred + q95
    coverage = np.mean((y_true >= lower) & (y_true <= upper))
    width = float(np.mean(upper - lower))
    print(f"[{zone}] CPI coverage(95%) = {coverage:.3f}, avg width = {width:.4f}")

    shap_idx = np.nan
    print(f"[{zone}] SHAP Stability removed from pipeline.")

    adv_auc = np.nan
    print(f"[{zone}] Adversarial validation skipped.")

    # ----------------------- Plotting -----------------------
    # 1) training history
    plt.figure(figsize=(8,4))
    plt.plot(history.history.get('loss', []), label='train_loss')
    plt.plot(history.history.get('val_loss', []), label='val_loss')
    plt.title(f"{zone} - Training Loss"); plt.legend()
    plt.tight_layout()
    plt.show()

    # 2) Actual vs Predicted (time series) — full test span (all test samples)
    plt.figure(figsize=(14,5))
    plt.plot(y_true_series.index, y_true_series.values, label='Actual', linewidth=1)
    plt.plot(pred_series.index, pred_series.values, label='Predicted', linewidth=1)
    plt.fill_between(pred_series.index, lower, upper, color='gray', alpha=0.25, label='CPI 95%')
    plt.title(f"{zone} - Actual vs Predicted"); plt.legend(); plt.xlabel("Datetime"); plt.ylabel(zone)
    plt.tight_layout()
    plt.show()

    # 3) True vs Pred scatter
    plt.figure(figsize=(6,5))
    plt.scatter(y_true, pred, alpha=0.4)
    mn, mx = np.min([y_true.min(), pred.min()]), np.max([y_true.max(), pred.max()])
    plt.plot([mn, mx], [mn, mx], color='red', linestyle='--')
    plt.xlabel("True"); plt.ylabel("Predicted")
    plt.title(f"{zone} - True vs Pred (R2={r2:.4f})"); plt.tight_layout(); plt.show()

    # 4) Residual histogram
    plt.figure(figsize=(8,3))
    plt.hist(y_true - pred, bins=40, color='gray')
    plt.title(f"{zone} - Residual Distribution (mean={np.mean(y_true-pred):.4f})"); plt.tight_layout(); plt.show()

    summary_results.append({
        "Zone": zone,
        "RMSE": float(rmse),
        "R2": float(r2),
        "PRR": float(prr),
        "CPI_Coverage": float(coverage),
        "CPI_Width": float(width),
        "SHAP_Stability": float(shap_idx) if shap_idx is not None else None,
        "Adversarial_AUC": float(adv_auc)
    })

# ----------------------- Final summary table -----------------------
res_df = pd.DataFrame(summary_results)
print("\n✅ Final Results Summary:")
print(res_df.round(4))

plt.figure(figsize=(12,4))
plt.subplot(1,3,1); plt.bar(res_df['Zone'], res_df['R2']); plt.ylim(0,1.05); plt.title("R² per Zone")
plt.subplot(1,3,2); plt.bar(res_df['Zone'], res_df['Adversarial_AUC']); plt.ylim(0,1.05); plt.title("Adversarial AUC per Zone")
plt.subplot(1,3,3); plt.bar(res_df['Zone'], res_df['PRR']); plt.ylim(0,1.05); plt.title("PRR per Zone")
plt.tight_layout(); plt.show()
```
